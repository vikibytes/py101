# Support Vector Machine Classifiers

Adapted from: https://pythonprogramming.net/linear-svc-example-scikit-learn-svm-python/

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm
```

```python
X = np.array([[1,2],
             [5,8],
             [1.5,1.8],
             [8,8],
             [1,0.6],
             [6,6],
             [4,2],
             [9,11]])

y = [0,1,0,1,0,1,0,1]
colors = ['red','green','blue','purple']
plt.scatter(X[:,0],X[:,1], c=y, cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```

```python
clf = svm.SVC(kernel='linear', C = 1.0)
clf.fit(X,y)
```

```python
w = clf.coef_[0]
print(w)

a = -w[0] / w[1]

xx = np.linspace(0,12)
yy = a * xx - clf.intercept_[0] / w[1]

h0 = plt.plot(xx, yy, 'k-', label="non weighted div")

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.legend()
plt.show()
```

```python
from sklearn import linear_model

logreg = linear_model.LogisticRegression(C=1e5) #C = inverse of regulisation constant

logreg.fit(X, y)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(6, 5))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot SVM Classifier decision boundary
w = clf.coef_[0]
print(w)

a = -w[0] / w[1]

xx = np.linspace(0,12)
yy = a * xx - clf.intercept_[0] / w[1]

h0 = plt.plot(xx, yy, 'k-')

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.legend()
plt.show()
```
